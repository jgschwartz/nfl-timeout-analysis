```python
import sys
sys.path.insert(0, '..')

from src.data.loader import load_seasons
from src.timeouts.detector import detect_early_timeouts
from src.timeouts.opportunities import compute_all_opportunities
from src.timeouts.scorer import score_timeouts, save_scores
from src.analysis.visualize import (
    plot_opportunity_cost_heatmap,
    plot_cost_distribution,
    plot_wasteful_by_coach,
    plot_cost_vs_timeouts_remaining,
    plot_best_missed_opportunity_timeline,
)
```

```python
SEASONS = list(range(2010, 2024))
pbp = load_seasons(SEASONS)
```

```python
early = detect_early_timeouts(pbp)
print(f'{len(early)} early timeouts detected')
```

```python
opps = compute_all_opportunities(pbp, early)
print(f'{len(opps)} opportunities computed')
```

```python
scores = score_timeouts(early, opps)
save_scores(scores)
print(scores['verdict'].value_counts())
```

```python
fig = plot_opportunity_cost_heatmap(scores)
fig.savefig('../reports/heatmap.png', dpi=150, bbox_inches='tight')
```

```python
for name, fig in [
    ('distribution', plot_cost_distribution(scores)),
    ('by_coach', plot_wasteful_by_coach(scores, min_samples=20)),
    ('vs_timeouts', plot_cost_vs_timeouts_remaining(scores)),
    ('timeline', plot_best_missed_opportunity_timeline(scores)),
]:
    fig.savefig(f'../reports/{name}.png', dpi=150, bbox_inches='tight')
```
